```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pp
```

```python
df = pd.read_csv("moi_data.csv")
df = df.rename(columns={'Radius' : 'Diameter'})


df
```

# Time Graphs and Table

```python
df_time = pd.DataFrame(df.iloc[:-1, [1,2]])

df_time.index += 1

df_time

```

```python
pp.scatter(df_time.index,df_time['Time (with)'],marker='o', label='Readings', s=100,alpha = 1,cmap = 'autumn', c = df_time['Time (with)'])
pp.errorbar(df_time.index,df_time['Time (with)'], yerr=0.01, fmt='none', ecolor='blue', capsize=10, label='Error Bars')
pp.style.use('classic')
cbar = pp.colorbar()
# linear regression model
coefficients = np.polyfit(df_time.index, df_time['Time (with)'], 1)
trendline = np.polyval(coefficients, df_time.index)

# Plot the trendline
pp.plot(df_time.index, trendline, color='magenta',linestyle='--', label='Trendline')
pp.title('Time Readings with Masses Attached')
pp.xlabel('Readings')
pp.ylabel('Time (s)')  
pp.legend()
pp.grid(True)
pp.xticks(df_time.index)
pp.ylim(7,11)
pp.xlim(1,10)
pp.show()
```

```python
pp.scatter(df_time.index,df_time['Time (without)'], marker='o', label='Readings', s=100,alpha = 1,cmap = 'autumn', c = df_time['Time (without)'])
pp.errorbar(df_time.index,df_time['Time (without)'], yerr=0.01, fmt='none', ecolor='blue', capsize=2, label='Error Bars')
cbar = pp.colorbar()
# linear regression model
coefficients = np.polyfit(df_time.index, df_time['Time (without)'], 1)
trendline = np.polyval(coefficients, df_time.index)

# Plot the trendline
pp.plot(df_time.index, trendline, color='magenta', linestyle='--', label='Trendline')
pp.title('Time Readings without Masses Attached')
pp.xlabel('Readings')
pp.ylabel('Time (s)')
pp.legend()
pp.grid(True)
pp.xticks(df_time.index)
pp.ylim(2,6)
pp.xlim(1,10)
pp.show()
```
